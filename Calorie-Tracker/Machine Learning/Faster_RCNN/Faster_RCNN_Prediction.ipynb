```python
from detecto import core, utils, visualize
from detecto.visualize import show_labeled_image
from detecto.utils import read_image
import pandas as pd
```

```python
def predict(image_dir):
    classes_list_df=pd.read_csv('Classes_List.csv')
    classes_list=list(classes_list_df['filename'])
    model = core.Model.load('model_weights_epochs_14_lr_0001_001.pth', classes_list)
    image=read_image(image_dir)
    predictions = model.predict(image)
    labels, boxes, scores = predictions
    show_labeled_image(image, boxes[0:1], labels[0:1])
    #print(scores)
```

```python
predict('aamras6.jpg')
```
