```python
import os
```

```python
def List_classes(source_dir):
    """
    Returns the folder names in source directory. 
    Source directory contains folders having images and annotations
    where each folder name will be the class name.
    """
    classes=next(os.walk(source_dir))[1]
    classes_list=list()
    for i in classes:
        classes_list.append(i)
    return classes_list
```
