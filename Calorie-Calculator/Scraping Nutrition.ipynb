```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
r=requests.get("https://www.calorie-charts.net/eggs/1015")
d=bs(r.text,"html.parser")
d
```

```python
print('Classes of each table:')
for table in d.find_all('table'):
    print(table.get('class'))
```

```python
table = d.find('table', class_='gridFDD')
table
```

```python
df = pd.DataFrame(columns=['Serving Size', 'Calories', 'Fat (g)', 'Carbohydrates (g)', 'Protein (g)'])

# Collecting Ddata
for row in table.find_all('tr'):    
    # Find all data for each column
    columns = row.find_all('td')
    if len(columns)==1:
        df = df.append({'Serving Size': columns[0].text.strip(),  'Calories': "", 'Fat (g)': "", 'Carbohydrates (g)': "", 'Protein (g)': ""}, ignore_index=True)
    elif(columns != [] and len(columns)==5):
        for i in columns:
            serving_size = columns[0].text.strip()
            if serving_size!="100"
            calories = columns[1].text.strip()
            fat = columns[2].text.strip()
            carb = columns[3].text.strip()
            protein = columns[4].text.strip()

        df = df.append({'Serving Size': serving_size,  'Calories': calories, 'Fat (g)': fat, 'Carbohydrates (g)': carb, 'Protein (g)': protein}, ignore_index=True)
```

```python
df.to_csv("scrape",sep=",")
```
