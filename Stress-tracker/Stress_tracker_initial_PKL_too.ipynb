```python
train = pd.read_csv('/content/drive/MyDrive/wesad-classification-hrv-train-dataset.csv')
test = pd.read_csv('/content/drive/MyDrive/wesad-classification-hrv-test-dataset.csv')
```

```python
conditions = dict(train['condition'].value_counts())
labels = list(conditions.keys())
counts = list(conditions.values())
plt.bar(labels,counts, color ='green',
        width = 0.4)
plt.show()
```

```python
le = preprocessing.LabelEncoder()
le.fit(train['condition'])
train['condition'] = le.transform(train['condition'])
test['condition'] = le.transform(test['condition'])
```

```python
train
```

```python
corr = train.corr()
```

```python
def feature_selection(correlation,threshold):
    selected_features = []
    for i in range(corr.shape[0]):
      if corr.iloc[i,35] > threshold:
        selected_features.append(train.iloc[:,i])
    return pd.DataFrame(selected_features).T
```

```python
#Selecting features having a correlation coefficient greater than 1
CORRELATION_THRESHOLD = 0.1
reduced_train = feature_selection(corr,CORRELATION_THRESHOLD)
```

```python
train
```

```python
cond = np.array(train['condition'])
```

```python
len(cond)
```

```python
reduced_train['condition'] = cond
```

```python
reduced_train
```

Split into X and y

```python
X_train = reduced_train.iloc[:,:-1]
y_train = reduced_train.iloc[:,-1]
```

```python
X_test = test[X_train.columns]
y_test = test['condition']
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_test.head()
```

Model Building

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
predicted_values = pd.DataFrame(le.inverse_transform(y_pred))
predicted_values.head(10)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
